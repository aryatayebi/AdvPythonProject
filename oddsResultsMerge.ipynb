```python
import pandas as pd
import numpy as np
```

```python
oddsDf = pd.read_csv('nbaOdds.csv')
resultsDf = pd.read_csv('nbaResults.csv')

oddsDf.drop(columns=['sport','gameState'],inplace=True)


print(oddsDf.shape)
print(resultsDf.shape)
```

```python
new_res = resultsDf[pd.isna(resultsDf.awayTeam)==False].copy()
resultsDf = resultsDf[pd.isna(resultsDf.awayTeam)].copy()
```

```python
resultsDf.columns
```

```python
new_res.drop(columns=['AwayDiff', 'AwayOutcome', 'AwayTeam', 'Date_Time', 'Game_State',
       'HomeDiff', 'HomeOutcome', 'HomeTeam', 'Sb_id', 'Score', 'Sport',
       'Total'],inplace=True)
resultsDf.drop(columns=['awayDiff', 'awayOutcome', 'awayTeam',
       'dateTime', 'gameState', 'homeDiff', 'homeOutcome', 'homeTeam', 'sbId',
       'score', 'sport', 'total'],inplace=True)
```

```python
resultsDf.columns
resultsDf.rename(columns = {'AwayDiff':'awayDiff', 'AwayOutcome':'awayOutcome', 'AwayTeam':'awayTeam', 
                            'Date_Time':'dateTime', 'Game_State':'gameState','HomeDiff':'homeDiff', 
                            'HomeOutcome':'homeOutcome', 'HomeTeam':'homeTeam', 'Sb_id':'sbId', 'Score':'score',
                            'Sport':'sport','Total':'total'}, inplace=True)
```

```python
oddsDf.rename(columns = {'AwayTeam':'awayTeam', 'Away_Odds':'awayOdds', 'Away_Spread':'awaySpread', 
                         'Away_Spread_Odds':'awaySpreadOdds', 'Book_Odds':'bookOdds', 'Consensus':'consensus', 
                         'Date_Time':'dateTime', 'Game_State':'gameState', 'HomeTeam':'homeTeam', 
                         'Home_Odds':'homeOdds','Home_Spread':'homeSpread', 'Home_Spread_Odds':'homeSpreadOdds',
                         'Over':'over', 'Over_Odds':'overOdds', 'Sb_id':'sbId',
                        'Sport':'sport', 'Under':'under', 'Under_Odds':'underOdds',}, inplace=True)
```

```python
resultsDf.shape
```

```python
new_res.shape
```

```python
resultsDf = pd.concat([resultsDf,new_res])
```

```python
### basic cleaning
oddsDf.drop(columns=['bookOdds','consensus','homeTeam','awayTeam','sbId','gameState','id','sport'],inplace=True)
resultsDf.drop(columns=['dateTime'],inplace=True)
```

```python
oddsDf.head(5)
```

```python
nbaData = resultsDf.merge(oddsDf, on='__id__')
```

```python
nbaData.to_csv('allNbaData.csv')
```

```python
nbaData
```

```python
playoffOdds = pd.read_csv('playoffOdds.csv')
playoffResults = pd.read_csv('playoffResults.csv')

playoffOdds.drop(columns=['bookOdds','consensus'],inplace=True)
playoffResults.drop(columns=['awayTeam','homeTeam','dateTime','__id__','gameState','sbId','sport'],inplace=True)

print(playoffOdds.shape)
print(playoffResults.shape)
```

```python
additonalDf = playoffOdds.merge(playoffResults, on='id')
```

```python
rawFinalDf = pd.concat([nbaData,additonalDf])
rawFinalDf.reset_index(inplace=True, drop=True)
rawFinalDf.shape
```

```python
finalDf = rawFinalDf.drop_duplicates(subset=['id'])
finalDf.reset_index(inplace=True, drop=True)
```

```python
finalDf.to_csv('finalNbaData.csv')
```
